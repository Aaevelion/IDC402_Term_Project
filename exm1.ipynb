```python
ops = 0

def add(a, b) :
    global ops
    c = a + b
    p = len(str(a))
    q = len(str(b))
    ops += (p + q)
    return c

def addlist(L) :
    if len(L) == 1 : return L[0]
    return add(L[0], addlist(L[1:]))

def return_ops(L) :
    global ops
    ops = 0
    addlist(L)
    return ops
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

num_x = 400
num_y = 400

x = range(num_x)
y = range(num_y)

X, Y = np.meshgrid(x,y)
Z = np.zeros((num_y, num_x))

for p in range(1, num_x):
   for k in range(1, num_y):
       Z[k, p] = return_ops([int('9'*p) for _ in range(k)])

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
# print(X.shape)
# print(Y.shape)
# print(Z.shape)
```
