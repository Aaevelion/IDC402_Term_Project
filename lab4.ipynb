```python
import numpy as np
from numpy import log
from scipy.special import factorial
import matplotlib.pyplot as plt
```

```python
from matplotlib import legend

x = np.linspace(15, 25, 100)

plt.plot(x, log(log(x)))
plt.plot(x, log(x))
plt.plot(x, log(x)**2)
plt.plot(x, x)
plt.plot(x, x**2)
plt.plot(x, x**log(x))
plt.plot(x, 2**x)
plt.plot(x, factorial(x))
plt.plot(x, x**x)
plt.legend(['log(log(x))', 'log(x)', 'log(x)^2', 'x', 'x^2', 'x^log(x)', '2^x', 'x!', 'x^x'])
plt.yscale('log')
plt.show()
```

```python
import time
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        # print(f"Took {end - start} seconds")
        return end - start, result
    return wrapper
```

```python
# def gcd(m, n) :
#     if m < n : return gcd(n, m)
#     if n > 0 : return gcd(n, m % n)
#     return m

@timer
def gcd(m, n) :
    i = 0
    if m < n : m, n = n, m
    while n > 0 : m, n, i = n, m % n, i+1
    return i #,m
```

```python
gcd(145, 108)
```

```python
iters = []
times = []
fibonacci = [1, 1]
n = 100

for _ in range(2, n):
    t, i = gcd(fibonacci[-1], fibonacci[-2])
    times.append(t)
    iters.append(i)
    fibonacci.append(fibonacci[-1] + fibonacci[-2])


fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.scatter(range(1, n-1), iters, c='r')
ax2.scatter(range(1, n-1), times, c='b')
ax1.set_xlabel('(n+1)th and (n+2)th Fibonacci number')
ax1.set_ylabel('Iterations for GCD', color='r')
ax2.set_ylabel('Wall clock time (s)', color='b')
ax1.legend(['Iterations fpr GCD'])
ax2.legend(['Wall clock time'], loc='upper right')
plt.show()
```
